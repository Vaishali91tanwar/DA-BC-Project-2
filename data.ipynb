```python
#Url to be queried
url="http://api.openweathermap.org/data/2.5/weather?appid="+weather_api_key+"&units=imperial&q="
```

```python
#Establishing the connection with a database
#db=client.weather_data
db = client.get_database("weather_data")
```

```python
###Adding to database

city_not_found=[]
city_found=[]
for city in us_cities:
    query_url=url+city
    response=requests.get(query_url)
    data=response.json()
    time.sleep(2)
    #print(data)
    try: 
        db.us.insert_one(data)
        city_found.append(city)  
    except Exception:
        print(f"City not found:{city}")
        city_not_found.append(city)
            
```
